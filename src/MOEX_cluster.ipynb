```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler


#MOEX_russia_company.csv
df = pd.read_csv('..\\data\\MOEX_russia_company2.csv', index_col=['Id'], error_bad_lines=False)


kmeans_data = df.iloc[:, 5:67]
features_value = kmeans_data.values

scaler = StandardScaler()
scaler.fit(kmeans_data)
scaled_features=scaler.transform(kmeans_data)
#print(scaled_features)

kmeans = KMeans(n_clusters=9, random_state=100)
kmeans.fit(scaled_features)
y_kmeans = kmeans.predict(scaled_features)
kmeans_data['cluster'] = y_kmeans


df['cluster'] = y_kmeans

df.to_csv('..\\data\\cluster_res.csv', sep=';', mode='a')
#print(df)
#
print(df.iloc[56])
print(df.iloc[70])
print(df[df['cluster']==0])
#58 PETROPAVLOVSK 72 Мечел

#model = KMeans(n_clusters=3)


```
